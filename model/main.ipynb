### Importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import sqlite3
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
cnx = sqlite3.connect("/home/kiuseh/Masaüstü/python/weatherPredictProject/model/dataBaseBucak.db")

data = pd.read_sql_query("SELECT * FROM DATABASE", cnx)
```

```python
data.head()
```

## data preprocessing

```python
le = LabelEncoder()
data["encoded_ruzgar_yonu"] = le.fit_transform(data["ruzgar_yonu"])
```

```python
data["encoded_ruzgar_yonu"].values
```

```python
data.head()
```

```python
data = data.rename(columns={"yagis_oranı": "yagis_orani"})
```

```python
X1 = data.drop(columns=["ruzgar_yonu", "uv_indeksi", "kontrol","yagis_orani"], axis=1)
y1 = data["yagis_orani"]
```

```python
print(y1)
```

## splitting dataset

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size = 0.2)
```

```python
model = RandomForestRegressor(random_state=42)
model.fit(X_train1, y_train1)
```

```python
y_pred1 = model.predict(X_test1)
```

```python
mae = mean_absolute_error(y_test1, y_pred1)
mse = mean_squared_error(y_test1, y_pred1)
rmse = np.sqrt(mse)
r2 = r2_score(y_test1, y_pred1)
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R²): {r2}")
```
